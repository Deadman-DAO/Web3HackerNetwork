```python
from pydriller import Repository
from datetime import datetime as dt
def introspect(obj):
  for func in [type, id, dir, vars, callable]:
        print("%s(%s):\t\t%s" % (func.__name__, introspect.__code__.co_varnames[0], func(obj)))


start = dt.now().timestamp()
commit_count = 0
file_count = 0
#for commit in Repository('./repos/enigmatt/new2Web3').traverse_commits():
for commit in Repository('./repos/oceanprotocol/contracts').traverse_commits():
    add_cnt = 0
    del_cnt = 0
    commit_count += 1
    this_file_count = 0
    for n in commit.modified_files:
        file_count += 1
        this_file_count += 1
        add_cnt += n.added_lines
        del_cnt += n.deleted_lines
    print(
        '{} modby {}, '
        ' in date {} {} {} {}'.format(
            commit.hash,
            commit.author.name,
            commit.committer_date,
            this_file_count,
            add_cnt,
            del_cnt
        )
    )
end = dt.now().timestamp() - start    
print(commit_count, 'commits', file_count, 'files', end, 'elapsed', end/file_count, 'seconds per file')
```
